```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup


browser = webdriver.Chrome(executable_path='chromedriver')
root = 'https://www.yourator.co'

tw_yourator = {}
tw_yourator['items'] = []
for page in range(1, 37):
    url = "https://www.yourator.co/companies?page=" + str(page)
    browser.get(url)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html5lib')
    
    table = soup.find('div', class_='row')
    for t in table:
        companyName = t.find('div', class_='y-card-content-title').get_text()
        domain = root + t.a['href']
#         description = t.find('div', class_='y-card-content-description').get_text().strip() # Simple One
        region = t.find_all('div', class_='y-card-content-icon-row')[0].get_text().strip()
        sector = t.find_all('div', class_='y-card-content-icon-row')[1].get_text().strip()
        # Get capital, and detail description
        browser.get(domain)
        html_source = browser.page_source
        soup_p = BeautifulSoup(html_source, 'html5lib')
        
        try:
            capital = soup_p.find('p', class_='basic-info__icon basic-info__icon--capital').get_text().strip()
            scale = soup_p.find('p', class_='basic-info__icon basic-info__icon--scale').get_text().strip()
        except:
            capital = '-'
            scale = '-'    
        try:
            hrefDivs = soup_p.find('div', class_='external-icons flex flex-wrap')
            links = hrefDivs.findAll('a')[0]['href'] 
            ahref = links if 'facebook' not in links else hrefDivs.findAll('a')[1]['href'] 
        except:
            ahref = domain
        
        real_domain = ahref
        description = soup_p.find('section', class_='content__area').get_text().strip().replace('\n', ' ').replace('\t', ' ').replace('\r', ' ')
        
        # Output Format
        obj = {}
        obj['companyName'] = companyName
        obj['domain'] = real_domain
        obj['description'] = description
        obj['region'] = region
        obj['sector'] = sector
        obj['capital'] = capital
        obj['scale'] = scale
        tw_yourator['items'].append(obj)
```

```python
print(tw_yourator)
# 複製後用下面這個網址直接 JSON 轉 csv 比較快
# https://json-csv.com/

import json

# 檔案超過 1MB，要拆檔案
n = len(tw_yourator['items'])
print(n)
tw_yourator1 = {}
tw_yourator1['items'] = tw_yourator['items'][:n//2]
tw_yourator2 = {}
tw_yourator2['items'] = tw_yourator['items'][n//2:]

# 輸出JSON檔案用
with open('tw_startup1.json', 'w', encoding='utf-8') as f:
    json.dump(tw_yourator1, f)
with open('tw_startup2.json', 'w', encoding='utf-8') as f:
    json.dump(tw_yourator2, f)
```
