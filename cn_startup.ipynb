```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.maigoo.com/news/525848.html'
headers = {'user-agent': 'Mozilla/5.0 (Macintosh Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.181 Safari/537.36'}
resp = requests.get(url, headers=headers)
resp.encoding = 'utf-8'
soup = BeautifulSoup(resp.text, 'html5lib')

table = soup.find_all('tr', class_='md_tr')
```

```python
import time

china_2019 = {}
china_2019['items'] = []

for i in table[1:]:
    j = i.find_all('td')
    companyName = j[1].get_text().strip()
    sector = j[2].get_text().strip()
    
    try:
        link = j[1].a['href']
        # Get into the link
        soup_p = requests.get(link, headers=headers)
        soup_p.encoding = 'utf-8'
        soup_p = BeautifulSoup(soup_p.text, 'html5lib')
        # For descript
        div = soup_p.find('div', class_='desc font16')
        string = ''
        for p in div.find_all('p'):
            string += p.text.strip()
        description = string
        # For addi info.
        ul = soup_p.find('ul', class_='c666').find_all('li')
        domain = ul[1].text.strip()
        founder = ul[2].text.strip()
        region = ul[3].text.strip()[4:]
        year = ul[4].text.strip()[5:]
        email = ul[5].text.strip()[5:]
        # Output data
        obj = {}
        obj['companyName'] = companyName
        obj['domain'] = domain
        obj['description'] = description
        obj['region'] = region
        obj['founder'] = founder
        obj['sector'] = sector
        obj['year'] = year
        obj['email'] = email
        china_2019['items'].append(obj)
    except:
        # Output data (Withoout Link)
        obj = {}
        obj['companyName'] = companyName
        obj['domain'] = '-'
        obj['description'] = '-'
        obj['region'] = '-'
        obj['founder'] = '-'
        obj['sector'] = sector
        obj['year'] = '-'
        obj['email'] = '-'
        china_2019['items'].append(obj)
    print(companyName)
```

```python
print(china_2019)
```
